# Cybercrime Incident Reports Analysis (2014-2024)

## Cumulative Reported Incidents by Month

The following analysis focuses on the cybercrime incidents reported in Vlaams-Brabant over the last decade (2014-2024). The data provided visualizes trends on a Radiant graph and includes monthly incident counts.

### Overview of Findings

Based on the data and the provided graph, here are the key insights:

1. **Peak Activity (October - December)**:
   - **Highest incident rates** are seen in the months of **October, November, and December**, which aligns with the end-of-year rush and higher online activity. Incidents sharply increase during this period, which could correlate with the rise in e-commerce, holiday shopping, and year-end financial transactions.
   - **November** is particularly noteworthy for its sharp spike, as seen in both the radiant chart and data records.

2. **Moderate Activity (July - August)**:
   - There is a **mid-year increase** in incidents, especially in **June and July**. This may be attributed to travel-related cybercrimes or mid-year financial frauds.
   - These months show steady but not extreme levels of cybercrime compared to the end of the year.

3. **Early-Year Decline (January - April)**:
   - The **first quarter of the year** (January through April) consistently shows a lower number of incidents. This period appears to have the **lowest activity** compared to other months. This post-holiday dip in cybercrime activity might result from less online shopping or decreased general cyber activity.

4. **Significant Incidents Data**:
   - The **total number of incidents** fluctuates between **87 and 150** incidents per month in 2014, and between **370 and 492** incidents per month in 2023, with noticeable increases from **March onward** and peaks towards the end of the year.

```python
# Rename the columns
df.columns = ['Date', 'Reported incident']
```

```python
# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python

# Create a line graph
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Create a line plot
plt.plot(df['Date'], df['Reported incident'], marker='o', linestyle='-', color='b', label='Reported Incidents')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Reported incident')
plt.title('Line Graph of Reported Incidents Over Time')

# Add a legend
plt.legend()

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python

# Extract month and year from the 'Date' column
df['YearMonth'] = df['Date'].dt.to_period('M')
```

```python
# Extract month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Group by 'Month' and aggregate the 'Reported incident' column
monthly_data = df.groupby('Month')['Reported incident'].sum().reset_index()

# Create a line graph
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Create a line plot
plt.plot(monthly_data['Month'], monthly_data['Reported incident'], marker='o', linestyle='-', color='b', label='Reported Incidents')

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Reported incident')
plt.title('Cumulative Reported Incidents by Month')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python

# Group by 'Month' and aggregate the 'Reported incident' column
monthly_data = df.groupby('Month')['Reported incident'].sum().reset_index()

# Create a radial chart
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# Convert months to radians for the polar plot
monthly_data['Month_rad'] = monthly_data['Month'] * (2 * np.pi / 12)

# Create a polar plot
ax = plt.subplot(111, polar=True)
ax.plot(monthly_data['Month_rad'], monthly_data['Reported incident'], marker='o', linestyle='-', color='b', label='Reported Incidents')

# Set the labels for each month
ax.set_xticks(monthly_data['Month_rad'])
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# Add labels and title
plt.title('Cumulative Reported Incidents by Month (Radial Chart)')

# Add a legend
plt.legend(loc='upper right')

# Show the plot
plt.show()
```

```python
# Group by 'Month' and aggregate the 'Reported incident' column
monthly_data = df.groupby('Month')['Reported incident'].sum().reset_index()

# Function to plot radial bar chart with color intensity
def plot_radial_chart(df, cmap_name='plasma'):
    labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    values = df['Reported incident']

    # Number of variables
    num_vars = len(labels)

    # Compute angle of each bar
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # The plot is a circle, so we need to "complete the loop"
    values = np.concatenate((values, [values[0]]))
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

    # Normalize the values for color mapping
    norm = plt.Normalize(df['Reported incident'].min(), df['Reported incident'].max())
    cmap = plt.get_cmap(cmap_name)

     # Draw bars with color intensity
    for angle, value, label in zip(angles[:-1], df['Reported incident'], labels):
        ax.bar(angle, value, color=cmap(norm(value)), edgecolor='black', linewidth=1, width=0.3)

    # Draw one axe per variable and add labels
    ax.set_yticklabels([])
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    plt.title('Cumulative Reported Incidents by Month 2014-2024', size=20, color='black', y=1.1)
    plt.show()

# Plot the radial chart
plot_radial_chart(monthly_data)
```
